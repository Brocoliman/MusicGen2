# Load LAKH MIDI

```python
import os
import midi
import matplotlib.pyplot as plt

os.chdir('/Users/jinghuang/Desktop/mgen/datasets/LAKH/')
```

```python
test_fn = 'ZZ_Top/Gimme_All_Your_Lovin.1.mid'

song = midi.read_midifile(test_fn)
song.make_ticks_abs()
tracks = []
for track in song:
    notes = [note for note in track if note.name == 'Note On']
    pitch = [note.pitch for note in notes]
    tick = [note.tick for note in notes]
    tracks += [tick, pitch]
plt.plot(*tracks)
plt.show()
```

```python
folders = os.listdir()
filenames = []
for folder in folders:
    if os.path.isfile(folder): continue
    folder_dir = os.listdir(folder)
    for file in folder_dir:
        filenames.append(os.path.join(folder, file))
filenames
```
